```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python
zipcodes = gpd.read_file('../data/zipcodes.geojson')
print(zipcodes.crs)
zipcodes.head()
```

```python
parks = gpd.read_file('../data/park_boudaries.geojson')
```

```python
print(parks.crs)
print(parks.shape)
parks.head()
```

```python
water = gpd.read_file('../data/water_fountains.geojson')
print(water.crs)
water.head()
```

```python
nashville = (36.16784, -86.77816)
park_map = folium.Map(location = nashville, zoom_start = 10)
folium.GeoJson(parks['geometry']).add_to(park_map)
#folium.GeoJson(zipcodes['geometry']).add_to(park_map)

for row_index, row_values in water.iterrows():
    loc = [row_values['lon'], row_values['lat']]
    pop = str(row_values['name']) + ' Location: ' + str(row_values['location']) + ' Dog: ' + str(row_values['dog'])
    icon=folium.Icon(color="blue",icon="tint", prefix='fa')
    
    marker = folium.Marker(
        location = loc, 
        popup = pop, icon = icon) 

    marker.add_to(park_map)
```
